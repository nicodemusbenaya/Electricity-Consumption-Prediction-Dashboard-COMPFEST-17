Engineer Fitur untuk kolom date

```python
# engineer fitur bedasarkan kolom date

# Fitur engineering pada kolom 'date'
# fitur yang dibuat harus sama persis untuk kedua dataset
# fitur ditaro di dataframe yang sudah dilakukan encoding
train_df_processed['year'] = train_df_processed['date'].dt.year
train_df_processed['month'] = train_df_processed['date'].dt.month
train_df_processed['day'] = train_df_processed['date'].dt.day
train_df_processed['dayofweek'] = train_df_processed['date'].dt.dayofweek
train_df_processed['dayofyear'] = train_df_processed['date'].dt.dayofyear
train_df_processed['weekofyear'] = train_df_processed['date'].dt.isocalendar().week.astype(int)
train_df_processed['quarter'] = train_df_processed['date'].dt.quarter
train_df_processed['is_month_start'] = train_df_processed['date'].dt.is_month_start.astype(int)
train_df_processed['is_month_end'] = train_df_processed['date'].dt.is_month_end.astype(int)
train_df_processed['is_year_start'] = train_df_processed['date'].dt.is_year_start.astype(int)
train_df_processed['is_year_end'] = train_df_processed['date'].dt.is_year_end.astype(int)
train_df_processed['is_weekend'] = (train_df_processed['date'].dt.dayofweek >= 5).astype(int)

test_df_processed['year'] = test_df_processed['date'].dt.year
test_df_processed['month'] = test_df_processed['date'].dt.month
test_df_processed['day'] = test_df_processed['date'].dt.day
test_df_processed['dayofweek'] = test_df_processed['date'].dt.dayofweek
test_df_processed['dayofyear'] = test_df_processed['date'].dt.dayofyear
test_df_processed['weekofyear'] = test_df_processed['date'].dt.isocalendar().week.astype(int)
test_df_processed['quarter'] = test_df_processed['date'].dt.quarter
test_df_processed['is_month_start'] = test_df_processed['date'].dt.is_month_start.astype(int)
test_df_processed['is_month_end'] = test_df_processed['date'].dt.is_month_end.astype(int)
test_df_processed['is_year_start'] = test_df_processed['date'].dt.is_year_start.astype(int)
test_df_processed['is_year_end'] = test_df_processed['date'].dt.is_year_end.astype(int)
test_df_processed['is_weekend'] = (test_df_processed['date'].dt.dayofweek >= 5).astype(int)

# Hapus kolom 'date' asli karena sudah diekstrak fiturnya
train_df_processed = train_df_processed.drop('date', axis=1)
test_df_processed = test_df_processed.drop('date', axis=1)
```

```python
# standarisasi untuk kolom beripe numerik untuk dataframe x

from sklearn.preprocessing import StandardScaler

# Pilih kolom numerik, kecuali kolom target 'electricity_consumption' and 'ID'
numerical_cols_train = train_df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Remove the target variable if it's still in the numerical columns list
if 'electricity_consumption' in numerical_cols_train:
    numerical_cols_train.remove('electricity_consumption')


# Inisialisasi StandardScaler
scaler = StandardScaler()

# Lakukan standarisasi pada kolom numerik di dataframe train_df_processed
train_df_processed[numerical_cols_train] = scaler.fit_transform(train_df_processed[numerical_cols_train])

# Lakukan hal yang sama pada dataframe test_df_processed
numerical_cols_test = test_df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Remove the 'ID' column if it's still in the numerical columns list (although it should be object type)
if 'ID' in numerical_cols_test:
    numerical_cols_test.remove('ID')

test_df_processed[numerical_cols_test] = scaler.transform(test_df_processed[numerical_cols_test]) # Use scaler.transform on test data

# Now you have train_df_processed and test_df_processed with standardized numerical features
# You might want to separate features and target again for model training
X_train_scaled = train_df_processed.drop(['electricity_consumption', 'ID'], axis=1)
y_train = train_df_processed['electricity_consumption']
X_test_scaled = test_df_processed.drop('ID', axis=1)
```

```python
# evaluasi kinerja model pada data yang belum pernah dilihat selama pelatihan untuk membantu mendeteksi overfitting dan memilih model terbaik sebelum melakukan prediksi akhir dengan Pemisahan Data Pelatihan dan Validasi (Train-Validation Split)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Pisahkan data pelatihan menjadi set pelatihan dan validasi
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_scaled, y_train, test_size=0.2, random_state=42
)

# Latih model (contoh: Linear Regression) pada data pelatihan yang dibagi
model = LinearRegression()
model.fit(X_train_split, y_train_split)

# Evaluasi model pada data validasi
y_val_pred = model.predict(X_val_split)

# Hitung Mean Squared Error (MSE) pada data validasi
mse_val = mean_squared_error(y_val_split, y_val_pred)
rmse_val = np.sqrt(mse_val)

print(f"RMSE pada set validasi: {rmse_val}")

# Bandingkan dengan performa pada data pelatihan untuk mendeteksi overfitting
y_train_pred_split = model.predict(X_train_split)
mse_train_split = mean_squared_error(y_train_split, y_train_pred_split)
rmse_train_split = np.sqrt(mse_train_split)

print(f"RMSE pada set pelatihan (split): {rmse_train_split}")


```

```python
# memeriksa outlier untuk kolom target

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.histplot(train_df['electricity_consumption'], bins=50, kde=True)
plt.title('Histogram of Electricity Consumption')
plt.xlabel('Electricity Consumption')
plt.ylabel('Frequency')
plt.show()

# Menggunakan Z-score untuk mendeteksi outlier (opsional, tergantung metode yang diinginkan)
from scipy import stats

z = np.abs(stats.zscore(train_df['electricity_consumption']))
threshold = 3 # Ambang batas Z-score, umum digunakan 2, 2.5, atau 3
outlier_indices = np.where(z > threshold)
print(f"Jumlah outlier yang terdeteksi menggunakan Z-score > {threshold}: {len(outlier_indices[0])}")
print("Indeks baris yang merupakan outlier:")
print(outlier_indices[0])

# Menggunakan IQR (Interquartile Range) untuk mendeteksi outlier
Q1 = train_df['electricity_consumption'].quantile(0.25)
Q3 = train_df['electricity_consumption'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = train_df[(train_df['electricity_consumption'] < lower_bound) | (train_df['electricity_consumption'] > upper_bound)]

print(f"Jumlah outlier yang terdeteksi menggunakan IQR (1.5 * IQR): {len(outliers_iqr)}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor # Jika XGBoost sudah diinstal
from lightgbm import LGBMRegressor # Jika LightGBM sudah diinstal
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

#  X_train_scaled, y_train, dan X_test_scaled

# 1. Definisikan Algoritma yang Akan Diuji

models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(random_state=42),
    'Lasso Regression': Lasso(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42, n_estimators=100), # n_estimators bisa di-tune
    'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=42, n_estimators=100), # n_estimators bisa di-tune
    'XGBoost Regressor': XGBRegressor(random_state=42, n_estimators=100, objective='reg:squarederror'), # Tambahkan objective
    'LightGBM Regressor': LGBMRegressor(random_state=42, n_estimators=100) # n_estimators bisa di-tune
}

# 2. Persiapkan Cross-Validation

# KFold untuk regresi: membagi data menjadi K 'lipatan'
kf = KFold(n_splits=5, shuffle=True, random_state=42) # 5-fold cross-validation

results = {}


# 3. Iterasi Melalui Setiap Algoritma dan Lakukan Cross-Validation

print("Memulai evaluasi model dengan K-Fold Cross-Validation...\n")
for name, model in models.items():
    print(f"Mengevaluasi {name}...")
    rmse_scores = []
    mae_scores = []
    r2_scores = []

    # Loop melalui setiap 'fold' dalam K-Fold Cross-Validation
    for train_index, val_index in kf.split(X_train_scaled):
        # Bagi data ke dalam pelatihan dan validasi untuk fold saat ini
        X_train_fold, X_val_fold = X_train_scaled.iloc[train_index], X_train_scaled.iloc[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        # Latih model pada data pelatihan fold
        model.fit(X_train_fold, y_train_fold)

        # Buat prediksi pada data validasi fold
        y_pred_fold = model.predict(X_val_fold)

        # Hitung metrik dan simpan
        rmse_scores.append(np.sqrt(mean_squared_error(y_val_fold, y_pred_fold)))
        mae_scores.append(mean_absolute_error(y_val_fold, y_pred_fold))
        r2_scores.append(r2_score(y_val_fold, y_pred_fold))

    # Hitung rata-rata metrik untuk semua fold
    avg_rmse = np.mean(rmse_scores)
    avg_mae = np.mean(mae_scores)
    avg_r2 = np.mean(r2_scores)

    results[name] = {
        'RMSE (Avg)': avg_rmse,
        'MAE (Avg)': avg_mae,
        'R2 (Avg)': avg_r2
    }
    print(f"  {name} - Avg RMSE: {avg_rmse:.4f}, Avg MAE: {avg_mae:.4f}, Avg R2: {avg_r2:.4f}\n")


# 4. Hasil Perbandingan

print("Ringkasan Hasil Evaluasi Model:")
results_df = pd.DataFrame(results).T # Transpose untuk tampilan lebih baik
results_df_sorted = results_df.sort_values(by='RMSE (Avg)', ascending=True) # Urutkan berdasarkan RMSE

print(results_df_sorted)

# 5. Pilih Model Terbaik dan Latih Ulang dengan Seluruh Data Pelatihan

best_model_name = results_df_sorted.index[0]
print(f"\nModel terbaik berdasarkan RMSE rata-rata validasi adalah: {best_model_name}")

# Inisialisasi model terbaik
best_model = models[best_model_name]

# Latih model terbaik pada SELURUH data pelatihan yang diskalakan
print(f"Melatih model terbaik ({best_model_name}) pada seluruh X_train_scaled dan y_train...")
best_model.fit(X_train_scaled, y_train)

print("\nModel terbaik sudah dilatih dan siap untuk prediksi akhir pada X_test_scaled.")
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
import numpy as np


# Inisialisasi parameter
k = 5  # jumlah fold
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Menyimpan skor
train_rmse_scores = []
val_rmse_scores = []

# K-Fold Cross Validation
for fold, (train_idx, val_idx) in enumerate(kf.split(X_train_scaled)):
    X_tr, X_val = X_train_scaled.iloc[train_idx], X_train_scaled.iloc[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    # Use the best_model trained on the entire dataset (from the previous cell)
    # Make predictions on the training and validation folds using the best_model
    y_tr_pred = best_model.predict(X_tr)
    y_val_pred = best_model.predict(X_val)

    # Calculate RMSE for the current fold
    train_rmse = np.sqrt(mean_squared_error(y_tr, y_tr_pred))
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))

    train_rmse_scores.append(train_rmse)
    val_rmse_scores.append(val_rmse)

    print(f"Fold {fold+1}: Train RMSE = {train_rmse:.4f}, Val RMSE = {val_rmse:.4f}")


# Hitung rata-rata dan selisih
avg_train_rmse = np.mean(train_rmse_scores)
avg_val_rmse = np.mean(val_rmse_scores)
rmse_gap = avg_val_rmse - avg_train_rmse
rel_gap_percent = (rmse_gap / avg_train_rmse) * 100

print("\n=== Ringkasan Cross-Validation ===")
print(f"Rata-rata Train RMSE: {avg_train_rmse:.4f}")
print(f"Rata-rata Validasi RMSE: {avg_val_rmse:.4f}")
print(f"Selisih RMSE: {rmse_gap:.4f} ({rel_gap_percent:.2f}%)")

# Analisis sederhana overfitting
if rel_gap_percent > 25:
    print("⚠️ Indikasi overfitting terdeteksi.")
else:
    print("✅ Model tampaknya tidak overfitting.")
```
